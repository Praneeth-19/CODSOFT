# TIC-TAC-TOE AI

Implement an AI agent that plays the classic game of Tic-Tac-Toe
against a human player. You can use algorithms like Minimax with
or without Alpha-Beta Pruning to make the AI player unbeatable.
This project will help you understand game theory and basic search
algorithms.

```python
import random

# Constants for the game
X = 'X'
O = 'O'
EMPTY = ' '

def print_board(board):
    for row in board:
        print(' '.join(row))
    print()

def empty_cells(board):
    return [(i, j) for i in range(3) for j in range(3) if board[i][j] == EMPTY]

def check_winner(board, player):
    # Check rows, columns, and diagonals
    for i in range(3):
        if all(board[i][j] == player for j in range(3)) or all(board[j][i] == player for j in range(3)):
            return True
    if all(board[i][i] == player for i in range(3)) or all(board[i][2 - i] == player for i in range(3)):
        return True
    return False

def game_over(board):
    return check_winner(board, X) or check_winner(board, O) or not any(EMPTY in row for row in board)

def minimax(board, depth, maximizing_player):
    if check_winner(board, X):
        return -1
    elif check_winner(board, O):
        return 1
    elif not any(EMPTY in row for row in board):
        return 0

    if maximizing_player:
        max_eval = float('-inf')
        for i, j in empty_cells(board):
            board[i][j] = O
            eval = minimax(board, depth + 1, False)
            board[i][j] = EMPTY
            max_eval = max(max_eval, eval)
        return max_eval
    else:
        min_eval = float('inf')
        for i, j in empty_cells(board):
            board[i][j] = X
            eval = minimax(board, depth + 1, True)
            board[i][j] = EMPTY
            min_eval = min(min_eval, eval)
        return min_eval

def get_best_move(board):
    best_eval = float('-inf')
    best_move = None
    for i, j in empty_cells(board):
        board[i][j] = O
        eval = minimax(board, 0, False)
        board[i][j] = EMPTY
        if eval > best_eval:
            best_eval = eval
            best_move = (i, j)
    return best_move

def main():
    board = [[EMPTY, EMPTY, EMPTY],
             [EMPTY, EMPTY, EMPTY],
             [EMPTY, EMPTY, EMPTY]]

    print_board(board)

    while not game_over(board):
        player_row = int(input("Enter your move row (0, 1, or 2): "))
        player_col = int(input("Enter your move column (0, 1, or 2): "))

        if board[player_row][player_col] == EMPTY:
            board[player_row][player_col] = X
        else:
            print("Cell already occupied. Try again.")
            continue

        print_board(board)

        if game_over(board):
            break

        print("AI is making a move...")
        ai_row, ai_col = get_best_move(board)
        board[ai_row][ai_col] = O

        print_board(board)

    if check_winner(board, X):
        print("You win!")
    elif check_winner(board, O):
        print("AI wins!")
    else:
        print("It's a draw!")

if __name__ == "__main__":
    main()
```
